```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#Defining a function to get the location of the localities
def get_location(localities):
    g = geocoder.arcgis('{}, Houston, US'.format(localities))
    get_latlng = g.latlng
    return get_latlng
```

```python
#Creating an empty list
co_ordinates = []
#Getting the co-ordinates of each locality using the function defined above
for i in houston_data["H Neighborhood"].tolist():
    co_ordinates.append(get_location(i)) 
print(co_ordinates)
```

```python
#Creating a dataframe from the list of location co-ordinates
co_ordinates_houston_data = pd.DataFrame(co_ordinates, columns=['Latitudes', 'Longitudes'])
```

```python
#Adding co-ordinates of localities to df dataframe
houston_data["Latitudes"] = co_ordinates_houston_data["Latitudes"]
houston_data["Longitudes"] = co_ordinates_houston_data["Longitudes"]
```
